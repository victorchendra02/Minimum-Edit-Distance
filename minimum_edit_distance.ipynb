```python
import numpy as np
import pandas as pd


def minimum_edit_distance(source: str, target: str):
    matrix = np.zeros((len(source)+1, len(target)+1), dtype=int)
    matrix[0] = [_ for _ in range(len(target)+1)]
    
    delete_cost = 1
    insert_cost = 1
    replace_cost = 2
    
    for __ in range(1, len(source)+1):
        matrix[__][0] = __
    
    source, target = "#" + source, "#" + target
    for i in range(1, len(matrix)):
        for j in range(1, len(matrix[0])):
            if source[i] == target[j]:
                cost = 0
            else:
                cost = replace_cost
                
            matrix[i][j] = min(
                matrix[i-1][j] + delete_cost,
                matrix[i][j-1] + insert_cost,
                matrix[i-1][j-1] + cost
            )

    return matrix


if __name__ == '__main__':
    source = "play"
    target = "plan"

    result = minimum_edit_distance(source, target)
    df = pd.DataFrame(result, index=list('#'+source), columns=list('#'+target))
    display(df)
    print(f"Minimum edit distance from '{source}' to '{target}' is {result[-1][-1]}")
```
